# Python Data Types

```num = 25```

- Here, 25 (an integer) is assigned to the num variable. So the data type of num is of the int class.

## Built-in data types in Python
- Numeric Types  : ```int, float, complex```
- String data types  : ```str```
- Sequence types  : ```list, tuple, range```
- Binary types  : ```bytes, bytearray, memoryview```
- Mapping data type  : ```dict```
- Boolean type  : ```bool```
- Set data types  : ```set, frozenset```

### Python Numeric Data Type
- int - holds signed integers of non-limited length (arbitrary precision).
- float - holds floating decimal points and it's accurate up to 15 to 17 decimal places.
- complex - holds complex numbers. (z = 5 + 2j)

````python
a = 200
b = 12.57467
c = 100+5j
print("The type of variable having value", a," is ", type(a))
print("The type of variable having value", b," is ", type(b))
print("The type of variable having value", c," is ", type(c))
````

* Float can also be scientific numbers with an "e" to indicate the power of 10.
    
    x = 25e4
    => 25x10^4

````python
x = 25e4
y = -67.6e100
print("variable x is type of ", type(x))
print("variable y is type of ", type(y))
````

# Python Numeric Data Type Conversion
- convert from int to float
a = float(x)
- convert from float to int
b = int(y)
- convert from int to complex
c = complex(x)

````python
a = 10.5
b = int(a)
print(a)
print(b)
print(type(a))
print(type(b))
````

````python
a = 10
b = float(a)
print(b)
print(type(b))
````
