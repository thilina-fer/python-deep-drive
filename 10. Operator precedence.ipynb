# Python Operator Precedence

**Operator Precedence** determines the order in which operations are processed in a mathematical or logical expression. Python follows a specific hierarchy to ensure calculations are consistent.

---

## 1. The Precedence Table (Highest to Lowest)

When multiple operators appear in the same expression, Python evaluates them in this order:

| Level | Operator | Description |
|:---:|:---|:---|
| **1** | `()` | **Parentheses** (Highest Priority) |
| **2** | `**` | Exponentiation (Power) |
| **3** | `+x`, `-x`, `~x` | Unary plus, Unary minus, Bitwise NOT |
| **4** | `*`, `/`, `//`, `%` | Multiplication, Division, Floor division, Modulus |
| **5** | `+`, `-` | Addition, Subtraction |
| **6** | `<<`, `>>` | Bitwise left and right shifts |
| **7** | `&` | Bitwise AND |
| **8** | `^` | Bitwise XOR |
| **9** | `|` | Bitwise OR |
| **10** | `==`, `!=`, `>`, `>=`, `<`, `<=`, `is`, `is not`, `in`, `not in` | **Comparisons**, Identity, and Membership |
| **11** | `not` | Logical NOT |
| **12** | `and` | Logical AND |
| **13** | `or` | Logical OR (Lowest Priority) |

---

## 2. Practical Examples

### Example A: Arithmetic Hierarchy
```python
# Expression: 10 + 5 * 2 ** 2
# 1. Exponentiation first: 2 ** 2 = 4
# 2. Multiplication next: 5 * 4 = 20
# 3. Addition last: 10 + 20 = 30
result = 10 + 5 * 2 ** 2
print(result) # Output: 30
