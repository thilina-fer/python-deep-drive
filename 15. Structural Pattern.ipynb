# Structural Pattern

The `match-case` statement is a powerful selection tool that checks both a variable's **value** and its **shape** (structure). 

* **Value Matching**: Checks if the data equals a specific value (e.g., `case 10:`).
* **Shape Matching**: Checks the structure of the data, such as:
    * [cite_start]Is it a `list`, `tuple`, or `dictionary`? [cite: 910, 912]
    * [cite_start]How many elements does it contain? [cite: 17, 85, 169]
    * [cite_start]Does it follow a specific pattern (e.g., a list with two strings)? [cite: 59, 131]

**Syntax:**
**Simplified Example:**
```python
# 'x' is the variable we are checking
x = 10 

match x:
    case 1:
        print("Value is one")
    case 10:
        print("Value is ten")
    case _:
        print("No match found") # The wildcard '_' acts like an 'else' block

````python
x = 10
match x:
    case 1:
        print("x is 1")
    case 10:
        print("x is 10")
    case _:
        print("x is something else")
````

````python
my_list = [1, 2, 3]
match my_list:
    case [x,y]:
        print(f"List has two elements: {x} and {y}")
    case [x,y,z]:  
        print(f"List has three elements: {x}, {y}, and {z}")
    case _:
        print("List has a different number of elements")
````
