# Loops

Loops allow us to execute a block of code repeatedly.

| Loop Type | Purpose | Keywords |
| :--- | :--- | :--- |
| **for** | Iterates over a sequence (List, Tuple, Range). | `for`, `in` |
| **while** | Repeats as long as a condition is True. | `while` |

**Key Points for Assignments:**
* [cite_start]**range(stop)**: Starts at 0 and goes up to (but not including) the stop value[cite: 125, 126].
* [cite_start]**Indentation**: All code inside the loop must be indented[cite: 463].
* **Infinite Loops**: In a `while` loop, ensure the condition eventually becomes `False`.

## For loop

A `for` loop is used to execute a block of code for each item in a sequence. This is called **Iteration**.

**Common Sequences used with `for`:**
1. **Lists / Tuples**: `[1, 2, 3]` or `("apple", "banana")`
2. **Strings**: `"Python"` (iterates through each character)
3. **Range**: `range(5)` (iterates through a sequence of numbers)

**Basic Syntax:**
```python
for variable in sequence:
    # Code to execute

````python
my_list = ["apple" , "banana", "cherry"]
for item in my_list:
    print(item)
````

````python
for i in range(7,100,7):
    print(i)
````
