# Control Flow Statement 

- Control Flow Statements are instructions that determine the order in which code is executed based on certain conditions or logic. Instead of simply running line-by-line from top to bottom, these statements allow a program to make decisions, skip sections of code, or repeat specific tasks

- Conditional Statements (Decision Making)These statements use Comparison and Logical operators to decide which path the program should take. 

    - **if statement**: Executes a block of code only if a condition is True.
    - **elif (else if)**: Checks another condition if the previous if was False.
    - **else**: Provides a default block of code to run if none of the conditions were met.
