# if , else Statements

1. The **if** StatementThe if statement is used to execute a block of code only if a specified condition is True.

2. The **else** StatementThe else statement is used to execute a block of code if the same condition is False

```python
x = 8

if x > 0 :
    print("x is positive")

print("This will always be printed")
```

```python
x = -8

if x > 0 :
    print("x is positive")
else:
    print("x is not positive")

print("This will always be printed")
```

```python
length_1  = float(input("Enter length_1 : "))
length_2 = float(input("Enter length_2 : "))

if (length_1 > 0 and length_2 > 0):
    hypotenuse = (length_1**2 + length_2**2)**0.5
    print("Hypotenuse length is : ", hypotenuse)
else:
    print("Error : Triangle sides must be positive numbers")
```

# if-elif-else Statements

Python uses the `if-elif-else` structure to execute different code blocks based on multiple conditions. 

* **if**: Evaluates the first condition. [cite_start]If it is `True`, the code block runs.
* **elif** (else if): Checks another condition only if the previous `if` was `False`. [cite_start]You can have multiple `elif` blocks[cite: 876].
* **else**: Executes a final block of code if **none** of the preceding conditions were `True`.

**Key Syntax Rules:**
1. Every condition line must end with a **colon (`:`)**.
2. [cite_start]The code inside each block must be **indented**[cite: 363, 473].
3. Once Python finds a `True` condition, it executes that block and skips the rest of the structure.

```python
num_1 = int(input("Enter num_1 : "))
num_2 = int(input("Enter num_2 : "))
num_3 = int(input("Enter num_3 : "))

if num_1 <= num_2 and num_1 <= num_3:
    smallest = num_1
    if num_2 <= num_3:
        mid, largest = num_2, num_3
    else:
        mid, largest = num_3, num_2
elif num_2 <= num_1 and num_2 <= num_3:
        smallest = num_2
        if num_1 <= num_3:
            mid, largest = num_1, num_3
        else:
            mid, largest = num_3, num_1
else:
        smallest = num_3
        if num_1 <= num_2:
            mid, largest = num_1, num_2
        else:
            mid, largest = num_2, num_1
print(num_1, num_2, num_3, "in sorted order is : ", smallest, mid, largest)
```
