# What are Operators?

Operators in Python are special symbols that perform operations on variables and values. They are used to performarithmetic operations, logical operations, and comparisons.

1. Arithmetic Operators 
2. Identity Operators
3. Assignment Operators 
4. Membership Operators
5. Comparison Operators 
6. Bitwise Operators
7. Logical Operators

```python
# 1. Arithmetic Operators
a = 10
b = 3
print(a + b)    # 13 (Addition)
print(a - b)    # 7  (Subtraction)
print(a * b)    # 30 (Multiplication)
print(a / b)    # 3.3333... (Division)
print(a % b)    # 1  (Modulus)
print(a ** 2)   # 100 (Exponentiation)
print(a // b)   # 3  (Floor division)

# 2. Identity Operators
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x
print(x is z)     # True 
print(x is y)     # False 
print(x is not y) # True

# 3. Assignment Operators
n = 5        # n = 5
n += 2       # n = n + 2 
n -= 1       # n = n - 1 
n *= 3       # n = n * 3 
n /= 2       # n = n / 2 

# 4. Membership Operators
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)      # True
print("orange" not in fruits)  # True

# 5. Comparison Operators
i = 10
j = 20
print(i == j)  # False 
print(i != j)  # True 
print(i > j)   # False 
print(i < j)   # True 
print(i >= 10) # True 

# 6. Bitwise Operators
# 5 = 0101 (Binary), 3 = 0011 (Binary)
print(5 & 3)  # 1  (AND: 0001)
print(5 | 3)  # 7  (OR: 0111)
print(5 ^ 3)  # 6  (XOR: 0110)
print(~5)     # -6 (NOT)
print(5 << 1) # 10 (Left Shift)

# 7. Logical Operators
age = 25
print(age > 18 and age < 30) # True 
print(age > 20 or age < 15)  # True 
print(not(age == 25))        # False 
```

- The arithmetic operators return the type of result dependson the type of operands, as below.
    - If either operand is a complex number, the result is converted to complex;
    - If either operand is a floating point number, the result is converted to floating point;

# Assignment Operators

- The assignment operators are used to assign values to variables.
- According to PEP 8 — Python’s official style guide: "Surround binary operators with a single space on both sides."

```python
# assignment
x = 5
print(x)

# add and assign
x = 5
x += 7
print(x)

# subtract and assign
x = 5
x -= 7
print(x)

# multiply and assign
x = 5
x *= 7
print(x)

# divide and assign
x = 5
x /= 7
print(x)

# modulo and assign
x = 5
x %= 7
print(x)

# exponentiation and assign
x = 5
x **= 7
print(x)

# floor division and assign
x = 5
x //= 7
print(x)
```
