# Python Lists

Lists are used to store multiple items in a single variable.
Lists are created using square brackets:

```
my_list = ["dog", "cat", "lion", "elephant"]
print(my_list )
```

- List items are indexed, the first item has index [0], the second item has index [1] etc. Python Lists are ordered.
- If you add new items to a list, the new items will be placed at the end of the list.
- Python lists are changeable. That means we can modify, add, and remove items in a list after it has been
created.


**Python list are also mutable.**
- Mutable means you can change the data or value it holds without changing its memory address.


````python
my_list = [1, 2, 3, 4, 5]
print(id(my_list))

my_list[0] = 10
print(my_list)
print(id(my_list))
````

## Lists allow duplicates. Since lists are indexed, lists can have items with the same value

````python
my_list = ["dog", "cat", "lion", "elephant", "cat"]
print(my_list )
````

## To determine how many items a list has, use the len() function

````python
my_list = ["dog", "cat", "lion", "elephant", "cat"]
print(my_list )
print(len(my_list))
````

## List items can be of any data type

````python
list1 = ["dog", "lion", "elephant"]
list2 = [1, 6, 13, 11, 3]
list3 = [True, False, False]
my_list = ["xyz", 22, True, 50, "cat"]

print(list1)
print(list2)
print(list3)
print(my_list)
````

## It is also possible to use the list() constructor when creating a new list.

````python
my_list = list(("cat", "lion", "dog")) # note the double round-brackets
print(my_list)
print(type(my_list))
````

# Access Items

### List items are indexed and you can access them by referring to the index number:

````python
my_list = ["dog", "cat", "lion", "elephant", "cat"]
print(my_list[0])
````

### Negative indexing means start from the end. -1 refers to the last item, -2 refers to the second last item etc.


````python
my_list = ["dog", "cat", "lion", "elephant", "cat"]
print(my_list[-1])
````

# Range of Indexes

- You can specify a range of indexes by specifying where to start and where to end the range.
- When specifying a range, the return value will be a new list with the specified items.
**The search will start at index 2 (included) and end at index 5 (not included).**

````python
my_list = ["dog", "cat", "lion", "elephant", "rat", "fish"]
print(my_list[2:5])
print(my_list[:4])
print(my_list[3:])
print(my_list[:])
````

# Change List Items

- To change the value of a specific item, refer to the index number:

````python
my_list = ["dog", "cat", "lion", "elephant", "rat", "fish"]
my_list[1] = "tiger"
print(my_list)

# Change the values "cat" and "lion" with the values "zebra" and "tiger":
my_list = ["dog", "cat", "lion", "elephant", "rat", "fish"]
my_list[1:3] = ["zebra", "tiger"]
print(my_list)
````

# Insert Items

- To insert a new list item, without replacing any of the existing values, we can use the insert() method.
- The insert() method inserts an item at the specified index:

````python
my_list = ["dog", "cat", "lion", "elephant", "rat", "fish"]
my_list.insert(2, "zebra")
print(my_list)
````

# Append Items

````python
my_list = ["dog", "cat", "lion", "elephant", "rat", "fish"]
my_list.append("zebra")
print(my_list)
````

# Extend List

- To append elements from another list to the current list, use the extend() method.
- Add the elements of your_list to my_list:

````python
my_list = ["dog", "cat", "lion"]
your_list = ["elephant", "rat", "fish"]
my_list.extend(your_list)
print(my_list)
````

# Remove Specified Item

- The remove() method removes the specified item.

````python
my_list = ["dog", "cat", "lion"]
my_list.remove("cat")
print(my_list)
````

# Remove Specified Index

- The pop() method removes the specified index.
- Add the elements of your_list to my_list:

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
my_list.pop(1)
print(my_list)
````

- The del keyword also removes the specified index:

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
del my_list[0]
print(my_list)
````

- The del keyword can also delete the list completely.

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
del my_list
print(my_list) #this will cause an error because "my_list" no longer exists
````

# Clear the List

- The clear() method empties the list. The list still remains, but it has no content.

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
my_list.clear()
print(my_list)
````

# Sort Lists

- List objects have a sort() method that will sort the list alphanumerically, ascending, by default: 
- Sort the list alphabetically:

- In alphanumeric sorting,
        
        0–9 come before A–Z.

        A–Z come before a–z (uppercase is smaller than lowercase in ASCII/Unicode order).

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
my_list.sort()
print(my_list)

my_list = [100, 50, 65, 82, 23]
my_list.sort()
print(my_list)
````

# Sort Descending

- To sort descending, use the keyword argument reverse = True:

````python
my_list = ["dog", "cat", "lion","rat","fish"]
my_list.sort(reverse = True)
print(my_list)

num_list = [109, 60, 45, 72, 14]
num_list.sort(reverse = True)
print(num_list)
````

# Case Insensitive Sort

- By default the sort() method is case sensitive, resulting in all capital letters being sorted before lower case
letters:
- So if you want a case-insensitive sort function, use str.lower as a key function:

````python
my_list = ["dog", "cat", "Lion", "Rat", "fish"]
my_list.sort(key = str.lower)
print(my_list)
````

# reverse list


- The reverse() method reverses the current sorting order of the elements.

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
my_list.reverse()
print(my_list)
````

# Copy a List

- You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.
- There are ways to make a copy, one way is to use the built-in List method copy().

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
new_list = my_list.copy()
print(new_list)
````

- Another way to make a copy is to use the built-in method list().

````python
my_list = ["dog", "cat", "lion", "rat", "fish"]
new_list = list(my_list)
print(new_list)
````

# Join Two Lists

- There are several ways to join, or concatenate, two or more lists in Python.
- One of the easiest ways are by using the + operator.

````python
list_1 = ["a", "b", "c"]
list_2 = [1, 2, 3]
list_3 = list_1 + list_2
print(list_3)
````

- you can use the extend() method, which purpose is to add elements from one list to another list:
- In the following example extend() method is used to add list2 at the end of list1:

````python
list_1 = ["a", "b" , "c"]
list_2 = [1, 2, 3]

list_1.extend(list_2)
print(list_1)
````
