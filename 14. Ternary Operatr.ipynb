# Ternary Operator

The ternary operator provides a shorthand way of writing an `if-else` statement. It returns one value if the condition is `True` and another if it is `False`.

**Format:**
`[on_true] if [expression] else [on_false]`

**Advantages:**
* Makes the code more concise and readable for simple conditions.
* Can be used directly inside a `print()` function or a variable assignment.

```python
x = int(input("Enter a number: "))
result = "Even" if x % 2 == 0 else "Odd"
print(result)
```

```python
x_1 = int(input("Enter a x_1: "))
x_2 = int(input("Enter a x_2: "))
y_1 = int(input("Enter a y_1: "))
y_2 = int(input("Enter a y_2: "))
slope = (y_2 - y_1) / (x_2 - x_1) if x_2 != x_1 else "Undefined"
print(f"Slope: {slope}")
```
